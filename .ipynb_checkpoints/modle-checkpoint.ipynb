```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
data.dropna(inplace=True)
data
```

```python
sns.pairplot(data[['Concentration','time','Integral','peakAbs1','MaxPeakAbs','MaxWaveLength']], diag_kind="kde")
```

```python
X = data.iloc[:,:-6]
y = data.iloc[:,0]
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X = std.fit_transform(X)
y = std.fit_transform(np.array(y).reshape(174,1))
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from keras import models,layers
model = models.Sequential()
model.add(layers.Dense(256,input_shape=(685,),activation='relu',kernel_initializer='normal'))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=32)
```

```python
from sklearn.metrics import r2_score
pred_train = model.predict(X_train)
print(r2_score(y_train,pred_train))

pred = model.predict(X_test)
print(r2_score(y_test,pred))
```

KFold Cross validation

```python
def baseline_model():
    model = models.Sequential()
    model.add(layers.Dense(256,input_shape=(685,),activation='relu',kernel_initializer='normal'))
    model.add(layers.Dropout(0.1))
    model.add(layers.Dense(64,activation='relu'))
    model.add(layers.Dense(1))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='Adam')
    return model
```

Evaluate Model


```python
from keras import models,layers
from scikeras.wrappers import KerasRegressor
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold

estimator = KerasRegressor(model=baseline_model, epochs=100, batch_size=20, verbose=0)
kfold = KFold(n_splits=7)
results = cross_validate(estimator, X, y, cv=kfold, scoring=('r2', 'neg_mean_squared_error'),return_train_score=True)
print(results['test_neg_mean_squared_error'])
print("train_r2_score:",results['train_r2'])
print("test_r2_score:",results['test_r2'])
```

```python
estimator.fit(X_train,y_train)
```

```python
pred = estimator.predict(X_test)
print(r2_score(y_test,pred))
```
